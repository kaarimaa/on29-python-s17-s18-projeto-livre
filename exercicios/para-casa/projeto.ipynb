```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_lesb = pd.read_csv("vivencia lesb.csv")
df_lesb
```

```python
df_lesb.info()
```

```python
df_lesb.isnull().sum()
```

```python
df_lesb.describe(include='all').T
```

```python
acesso_saude = df_lesb['Acesso à Saúde\nAtualmente, qual sua forma de acesso á saúde?'].value_counts()
acesso_saude
```

```python
# Filtrar pessoas que sofreram discriminação
df_discriminacao = df_lesb[df_lesb['Discriminação\nVocê já enfrentou discriminação ou preconceito por ser lésbica?'] == 'Sim']
discriminacao_raca = df_discriminacao['Como se identifica racial / etnicamente'].value_counts()
discriminacao_escolaridade = df_discriminacao['Educação\nEscolaridade'].value_counts()
discriminacao_regiao = df_discriminacao['Local de Residência\nRegião'].value_counts()
discriminacao_raca, discriminacao_escolaridade, discriminacao_regiao
```

```python
## Distribuição de escolaridade
escolaridade = df_lesb['Educação\nEscolaridade'].value_counts()
perfil_alta_escolaridade = df_lesb[df_lesb['Educação\nEscolaridade'] == df_lesb['Educação\nEscolaridade'].max()]
idade_alta_escolaridade = perfil_alta_escolaridade['Idade'].value_counts()
renda_alta_escolaridade = perfil_alta_escolaridade['Renda\nSua faixa de renda mensal pessoal (per capta)'].value_counts()
escolaridade, idade_alta_escolaridade, renda_alta_escolaridade
```

```python
# Visualização do acesso à saúde
plt.figure(figsize=(10, 6))
sns.countplot(data=df_lesb, x='Acesso à Saúde\nAtualmente, qual sua forma de acesso á saúde?', palette="flare")
plt.title('Acesso à Saúde')
plt.xlabel('Forma de Acesso')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()

# Visualização do perfil que mais sofre discriminação por raça
plt.figure(figsize=(10, 6))
sns.countplot(data=df_discriminacao, x='Como se identifica racial / etnicamente', palette='Set2')
plt.title('Distribuição de Discriminação por Raça')
plt.xlabel('Raça')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()

# Visualização do perfil com maior nível de escolaridade
plt.figure(figsize=(10, 6))
sns.countplot(data=perfil_alta_escolaridade, x='Idade', palette="magma")
plt.title('Distribuição de Idade para Perfil com Maior Nível de Escolaridade')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```
